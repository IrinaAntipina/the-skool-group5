```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_correct = pd.read_excel(
    "data/ek_1_utbet_statliga_medel_utbomr.xlsx",
    sheet_name='Utbetalda statliga medel',
    header=5
)
df_correct = df_correct.set_index('Utbildningsområde')

top4 = df_correct[2024].sort_values(ascending=False).head(4).index.tolist()

years = df_correct.columns.astype(str).tolist()
bar_width = 0.2
x = np.arange(len(years))

colors = ['#4B4B4B', '#7D7D7D', '#A9A9A9', '#D3D3D3']

plt.figure(figsize=(12,7))

for i, area in enumerate(top4):
    values = df_correct.loc[area].values
    plt.bar(x + i*bar_width, values, width=bar_width, label=area, color=colors[i])

plt.xticks(x + bar_width, years, rotation=45)
plt.xlabel('År')
plt.ylabel('Utbetalda statliga medel (miljoner kronor)')
plt.title('Utveckling av statliga medel 2012-2024 för topp 4 utbildningsområden')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df.columns = df.columns.astype(str)

year = "2024"
df_year = df[["Utbildningsområde", year]].dropna()
df_year = df_year.sort_values(by=year, ascending=True)

plt.figure(figsize=(12, 8))
bars = plt.barh(df_year["Utbildningsområde"], df_year[year], color="black")

for bar in bars:
    plt.text(
        bar.get_width() + 5,
        bar.get_y() + bar.get_height() / 2,
        f"{bar.get_width():.0f}",
        va="center",
        fontsize=10,
        color="black",
    )

target_label = "Ekonomi, administration och försäljning"
target_index = df_year.index[df_year["Utbildningsområde"] == target_label][0]

target_bar = bars[target_index]
x = target_bar.get_width()
y = target_bar.get_y() + target_bar.get_height() / 2

plt.annotate(
   "Endast 1 miljon skillnad i statsbidrag",
   xy=(x + 800, y + 12.5),
   xytext=(x + 2000, y + 11),
   arrowprops=dict(
       arrowstyle="-|>",
       connectionstyle="arc3,rad=0.1",
       color="black",
       lw=1.5
   ),
   fontsize=12,
   ha="left",
   va="center"
)

plt.title(f"Utbetalda statliga medel per utbildningsområde år {year}",
          fontsize=16, fontweight="bold")
plt.xlabel("Miljoner kronor", fontsize=12)
plt.ylabel("")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()

plt.show()
```
