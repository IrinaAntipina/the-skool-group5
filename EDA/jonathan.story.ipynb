```python
import pandas as pd 

df = pd.read_excel("data/ek_1_utbet_statliga_medel_utbomr.xlsx", sheet_name="Utbetalda statliga medel", skiprows=5, skipfooter=4)
df 
```

```python
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_long = df.melt(id_vars=["Utbildningsområde"], var_name="År", value_name="Statsbidrag")

df_long["År"] = pd.to_numeric(df_long["År"])

plt.figure(figsize=(15, 7))

sns.lineplot(data=df_long, x="År", y="Statsbidrag", hue="Utbildningsområde", marker="o")

plt.title("Utveckling av utbetalda statliga medel per utbildningsområde", fontsize=16)
plt.ylabel("Utbetalda medel (miljoner kr)", fontsize=14)
plt.xlabel("År", fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title="Utbildningsområde", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()

plt.show()
```
