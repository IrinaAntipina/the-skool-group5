```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_correct = pd.read_excel(
    "data/ek_1_utbet_statliga_medel_utbomr.xlsx",
    sheet_name='Utbetalda statliga medel',
    header=5
)
df_correct = df_correct.set_index('Utbildningsområde')

# Ta bort "Totalt" kolumnen om den finns
df_correct = df_correct.drop(columns=['Totalt'], errors='ignore')

top4 = df_correct[2024].sort_values(ascending=False).head(4).index.tolist()

years = df_correct.columns.astype(str).tolist()
bar_width = 0.2
x = np.arange(len(years))

colors = ['#4B4B4B', '#7D7D7D', '#A9A9A9', '#D3D3D3']

plt.figure(figsize=(12,7))

for i, area in enumerate(top4):
    values = df_correct.loc[area].values
    plt.bar(x + i*bar_width, values, width=bar_width, label=area, color=colors[i])

plt.xticks(x + bar_width, years, rotation=45)
plt.xlabel('År')
plt.ylabel('Utbetalda statliga medel (miljoner kronor)')
plt.title('Utveckling av statliga medel 2012-2024 för topp 3 utbildningsområden')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df.columns = df.columns.astype(str)

years = [str(year) for year in range(2012, 2025)]
df_areas = df.set_index('Utbildningsområde')[years].dropna()

df_percent = df_areas.div(df_areas.sum(axis=0), axis=1) * 100
df_percent = df_percent.drop("Totalt", errors='ignore')

top_n = 6
df_percent['mean'] = df_percent.mean(axis=1)
df_percent_sorted = df_percent.sort_values('mean', ascending=False)
top_categories = df_percent_sorted.head(top_n).drop(columns='mean')
other_categories = df_percent_sorted.iloc[top_n:].drop(columns='mean').sum().to_frame().T
other_categories.index = ["Övriga områden"]
df_final = pd.concat([top_categories, other_categories])

fig, ax = plt.subplots(figsize=(16, 10))

box = ax.get_position()
ax.set_position([box.x0 + 0.35, box.y0, box.width - 0.35, box.height])

width = 0.10
x = np.arange(len(years))

colors = ['#1a1a1a', '#333333', '#4d4d4d', '#666666', '#808080', '#999999', '#b3b3b3']

for i, area in enumerate(df_final.index):
    ax.bar(x + i * width, df_final.loc[area], width, label=area, color=colors[i], edgecolor='black')

ax.set_title('Procentuell fördelning av statliga medel över utbildningsområden (2012–2024)', 
             fontsize=18, fontweight='bold')
ax.set_xlabel('År', fontsize=14)
ax.set_ylabel('Procentuell fördelning (%)', fontsize=14)
ax.set_xticks(x + width * (len(df_final.index)-1)/2)
ax.set_xticklabels(years, rotation=45)

leg = ax.legend(title='Utbildningsområde', bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=11, title_fontsize=12)

for patch in leg.get_patches():
    patch.set_width(14)
    patch.set_height(14)

for text in leg.get_texts():
    text.set_fontsize(11)

ax.grid(axis='y', linestyle='--', alpha=0.6)

left = 1.02
bottom = 0.15
width = 0.32
height = 0.55

ax_table = fig.add_axes([left, bottom, width, height])
ax_table.axis('off')

cell_text = []
for area in df_final.index:
    row = [f"{df_final.loc[area, year]:.1f}%" for year in years]
    cell_text.append(row)

table = ax_table.table(cellText=cell_text,
                       rowLabels=df_final.index,
                       colLabels=years,
                       loc='center',
                       cellLoc='center',
                       rowLoc='center',
                       edges='closed')

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 2)

for key, cell in table.get_celld().items():
    cell.get_text().set_fontfamily('monospace')
    cell.get_text().set_color('black')
    if key[0] == 0 or key[1] == -1:
        cell.set_text_props(weight='bold')
        cell.set_facecolor('#bfbfbf')
    else:
        cell.set_facecolor('#ffffff')

plt.subplots_adjust(left=0.005, right=0.73, top=0.97, bottom=0.1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df.columns = df.columns.astype(str)

year = "2024"
df_year = df[["Utbildningsområde", year]].dropna()
df_year = df_year.sort_values(by=year, ascending=True)

plt.figure(figsize=(12, 8))
bars = plt.barh(df_year["Utbildningsområde"], df_year[year], color="black")

for bar in bars:
    plt.text(
        bar.get_width() + 5,
        bar.get_y() + bar.get_height() / 2,
        f"{bar.get_width():.0f}",
        va="center",
        fontsize=10,
        color="black",
    )

target_label = "Ekonomi, administration och försäljning"
target_index = df_year.index[df_year["Utbildningsområde"] == target_label][0]

target_bar = bars[target_index]
x = target_bar.get_width()
y = target_bar.get_y() + target_bar.get_height() / 2

plt.annotate(
   "Endast 1 miljon skillnad i statsbidrag",
   xy=(x + 800, y + 12.5),
   xytext=(x + 2000, y + 11),
   arrowprops=dict(
       arrowstyle="-|>",
       connectionstyle="arc3,rad=0.1",
       color="black",
       lw=1.5
   ),
   fontsize=12,
   ha="left",
   va="center"
)

plt.title(f"Utbetalda statliga medel per utbildningsområde år {year}",
          fontsize=16, fontweight="bold")
plt.xlabel("Miljoner kronor", fontsize=12)
plt.ylabel("")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()

plt.show()
```
