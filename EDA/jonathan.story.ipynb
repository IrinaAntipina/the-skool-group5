```python
import pandas as pd 

df = pd.read_excel("data/ek_1_utbet_statliga_medel_utbomr.xlsx", sheet_name="Utbetalda statliga medel", skiprows=5, skipfooter=4)
df 
```

```python
print(df.head())
```

```python
import matplotlib.pyplot as plt

year = "Värde"

df_year = df[["Utbildningsområde", year]].dropna()

df_year = df_year.sort_values(by=year, ascending=True)

plt.figure(figsize=(12, 8))

bars = plt.barh(df_year["Utbildningsområde"], df_year[year], color="black")

for bar in bars:
    plt.text(
        bar.get_width() + 5,
        bar.get_y() + bar.get_height() / 2,
        f"{bar.get_width():.0f}",
        va="center",
        fontsize=10,
        color="black",
    )

index_data_it = df_year.index[df_year["Utbildningsområde"] == "Data/IT"][0]
index_teknik = df_year.index[df_year["Utbildningsområde"] == "Teknik och tillverkning"][0]

x_data_it = df_year.loc[index_data_it, year]
x_teknik = df_year.loc[index_teknik, year]

y_data_it = bars[index_data_it].get_y() + bars[index_data_it].get_height() / 2
y_teknik = bars[index_teknik].get_y() + bars[index_teknik].get_height() / 2

plt.annotate(
    "3 högst statliga utbetalningar",
    xy=(x_data_it, y_data_it),
    xytext=(x_teknik + 30, y_teknik),
    arrowprops=dict(
        arrowstyle='-|>',
        connectionstyle="arc3,rad=0.3",
        color="black",
        lw=2
    ),
    fontsize=12,
    ha="left",
    va="center"
)

plt.title(f"Utbetalda statliga medel per utbildningsområde år {year}", fontsize=16, fontweight="bold")
plt.xlabel("Miljoner kronor")
plt.ylabel("")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
