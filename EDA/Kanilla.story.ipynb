Visualization 1: Geografisk fördelning av ansökningar 2025

## Visualisering 1: Geografisk fördelning av ansökningar för YH-kurser 2025
Denna graf visar hur ansökningar för YH-kurser är fördelade över Sveriges städer eller regioner baserat på antalet inkomna ansökningar. Detta hjälper The Skool att identifiera var aktivitet är koncentrerad och var det finns luckor.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Check if file exists
file_path = "data/inkomna-ansokningar-2025-for-kurser.xlsx"
if not os.path.exists(file_path):
    print(f"Error: File '{file_path}' not found. Please check the file path.")
else:
    try:
        # Load data from Lista ansökningar
        df = pd.read_excel(
            file_path,
            sheet_name="Lista ansökningar",
            header=0
        )

        # Display basic info
        print("\nColumns in DataFrame:", df.columns.tolist())
        print("\nFirst few rows of DataFrame:")
        print(df.head())

        # Filter out rows where Kommun is "Se Lista flera kommuner"
        df_geo = df[df["Kommun"] != "Se \"Lista flera kommuner\""]

        # Aggregate applications by Kommun
        df_geo = df_geo.groupby("Kommun")["Sökt antal platser 2025"].sum().reset_index()
        df_geo = df_geo.sort_values(by="Sökt antal platser 2025", ascending=False).head(10)

        # Plot bar
        plt.figure(figsize=(12, 6))
        plt.bar(df_geo["Kommun"], df_geo["Sökt antal platser 2025"], color="skyblue")
        plt.title("Topp 10 kommuner med flest sökta platser för YH-kurser 2025", fontsize=14, pad=15)
        plt.xlabel("Kommun", fontsize=12)
        plt.ylabel("Antal sökta platser 2025", fontsize=12)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig("geo_distribution_2025.png", dpi=300)
        plt.show()

    except Exception as e:
        print(f"Error reading file or processing data: {str(e)}")
```

### Analys
- **Vad visar grafen?** Grafen visar de 10 städerna med flest inkomna ansökningar för YH-kurser 2025. Stockholm och Göteborg dominerar, medan mindre städer som Umeå har färre ansökningar.
- **Insikter**: Koncentrationen av ansökningar i storstäder indikerar hög konkurrens, medan mindre regioner är underrepresenterade. Detta skapar möjligheter för The Skool att ansöka om kurser i dessa områden.
- **Rekommendationer**: The Skool bör överväga att rikta ansökningar till underrepresenterade regioner, t.ex. norra Sverige, för att möta lokala behov och minska konkurrens.

Visualisering 2: Aktivitet hos utbildningsanordnare för YH-kurser 2025
# Denna graf visar de mest aktiva utbildningsanordnarna baserat på totalt antal sökta platser 2025, eftersom data om godkända ansökningar saknas.


```python
try:
    # Aggregate applications by Anordnare namn
    df_anordnare = df.groupby("Anordnare namn")["Sökt antal platser 2025"].sum().reset_index()
    df_anordnare = df_anordnare.sort_values(by="Sökt antal platser 2025", ascending=False).head(10)

    # Plot bar
    plt.figure(figsize=(12, 6))
    plt.bar(df_anordnare["Anordnare namn"], df_anordnare["Sökt antal platser 2025"], color="lightgreen")
    plt.title("Topp 10 utbildningsanordnare efter sökta platser 2025", fontsize=14, pad=15)
    plt.xlabel("Utbildningsanordnare", fontsize=12)
    plt.ylabel("Antal sökta platser 2025", fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig("anordnare_activity_2025.png", dpi=300)
    plt.show()

except Exception as e:
    print(f"Error processing anordnare activity: {str(e)}")
```


### Analys
- **Vad visar grafen?** Grafen visar de 10 utbildningsanordnare med flest sökta platser 2025. IHM Business School AB Göteborg och andra stora aktörer dominerar.
- **Insikter**: Höga volymer indikerar stark närvaro och konkurrens. The Skool kan dra nytta av att analysera dessa aktörers strategier.
- **Rekommendationer**: The Skool bör överväga partnerskap med framgångsrika anordnare och fokusera på nischade kurser för att sticka ut.


# ## Visualisering 3: Populäritet för utbildningsområden baserat på sökta platser 2025
Denna graf visar antalet sökta platser per utbildningsområde, med fokus på Ekonomi, administration och försäljning.

```python
try:
    # Aggregate applications by utbildningsområde
    df_areas = df.groupby("Sökt utbildningsområde")["Sökt antal platser 2025"].sum().reset_index()
    df_areas = df_areas.sort_values(by="Sökt antal platser 2025", ascending=True)

    # Plot horizontal bar
    plt.figure(figsize=(12, 8))
    bars = plt.barh(df_areas["Sökt utbildningsområde"], df_areas["Sökt antal platser 2025"], color="black")
    for bar in bars:
        plt.text(
            bar.get_width() + 5,
            bar.get_y() + bar.get_height() / 2,
            f"{bar.get_width():.0f}",
            va="center",
            fontsize=10,
            color="black",
        )

    # Add annotation for Ekonomi, administration och försäljning
    target_label = "Ekonomi, administration och försäljning"
    if target_label in df_areas["Sökt utbildningsområde"].values:
        target_index = df_areas.index[df_areas["Sökt utbildningsområde"] == target_label][0]
        target_bar = bars[target_index]
        x = target_bar.get_width()
        y = target_bar.get_y() + target_bar.get_height() / 2
        plt.annotate(
            "Hög popularitet med liten skillnad",
            xy=(x + 10, y + 0.5),
            xytext=(x + 50, y + 1),
            arrowprops=dict(arrowstyle="-|>", connectionstyle="arc3,rad=0.1", color="black", lw=1.5),
            fontsize=12,
            ha="left",
            va="center"
        )

    plt.title("Sökta platser per utbildningsområde 2025", fontsize=16, fontweight="bold")
    plt.xlabel("Antal sökta platser 2025", fontsize=12)
    plt.ylabel("")
    plt.grid(axis="x", linestyle="--", alpha=0.7)
    plt.tight_layout()
    plt.savefig("popular_areas_2025.png", dpi=300)
    plt.show()

except Exception as e:
    print(f"Error processing areas: {str(e)}")
```
